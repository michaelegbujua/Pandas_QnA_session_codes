```python
#dummy variables are also known as indicator variables
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain') #source: kaggle data online
```

```python
train.head()
```

```python
#create a new column called 'sex_male'
train['sex_male'] = train.Sex.map({'female': 0, 'male': 1})
```

```python
train.head()
```

```python
#you can achieve concise result with get_dummies() method
pd.get_dummies(train.Sex)
```

```python
#to remove the female column, maybe you just want the male column
pd.get_dummies(train.Sex).iloc[ :, 1:]
```

```python
train.Embarked.value_counts()
```

```python
#to set a prefix as a column name :
pd.get_dummies(train.Sex, prefix ='Sex').iloc[:, 1:]
```

```python
pd.get_dummies(train.Embarked, prefix = 'Embarked')
```

```python
embarked_dummies = pd.get_dummies(train.Embarked, prefix = 'Embarked').iloc[:, 1:]
embarked_dummies
```

```python
#to combine the initial and the created dataframes
train = pd.concat([train, embarked_dummies], axis = 1) 
```

```python
train.head()
```

```python
#alternatively 
pd.get_dummies(train, columns = ['Sex', 'Embarked'])
```
